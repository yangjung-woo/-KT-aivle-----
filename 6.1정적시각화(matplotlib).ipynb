# 정적 시각화 실습 - matplotlib 활용

pyplot 의 plot

```python
# 필요 라이브러리 로딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 브라우저 내부(inline)에 바로 그려주도록 하는 코드 
%matplotlib inline
```

```python
#정규분포에서 난수를 추출
np.random.seed(0)

x =np.arange(50)
y=np.random.randn(50)

```

## 기본그래프 그리기 
1. figure을 만들고 --> plt.figure()
2. figure내 그래프를 그리고 -->  plt.plot(x,y)
3. 그래프를 출력 --> plt.show()

```python
# 기본적인 그래프 그리기 
plt.figure()
plt.plot(x,y)
plt.show()
```

## Figure 크기 설정 & 축 제목 표기

```python
# 타이틀과 x축, y 축 레이블 생성

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.xlabel('Period (t)')
plt.ylabel('Percent change')
plt.title('Change of Interest Rate Over Last 50 years')
plt.savefig('interest_rate.png')
plt.show()
```

## 그래프에 데이터 추가 

```python
np.random.seed(2)

z= np.random.randn(50)
```

```python
# 타이틀과 x축, y 축 레이블 생성

plt.figure(figsize=(10,5))
plt.plot(x,y, label = 'Korea')
plt.plot(x,z, label = 'USA')
plt.xlabel('Period (t)')
plt.ylabel('Percent change')
plt.title('Change of Interest Rate Over Last 50 years')
plt.savefig('interest_rate.png')
plt.legend()
plt.show()
```

## 여러 그래프 그리기 1: subplot

matplotlib의 pyplot이 제공하는 subplot 메서드는 여러개의 그래프를 하나의 그림에 나타나도록 함

subplot은 그림의 위치와 순서에 관련한 3개의 값을 설정해줘야 합니다. 따라서 최종적으로 우리가 사용하고자 하는 subplot의 커맨드는  plt.subplot(행,열,인덱스)


```python
a = np.linspace(0.0,3.0,31)
```

```python
sin = np.sin(2*np.pi*a)

decaying_sin = np.sin(2*np.pi*a) * np.exp(-a)
```

```python
plt.subplot(2,1,1)
plt.plot(a,decaying_sin,'-g')
plt.title('Sin Graph Approaching Zero')

plt.subplot(2,1,2)
plt.plot(a,sin,'o-')
plt.title('Sin Graph')

plt.show()
```

### sharex: x축 공유

```python
plt.subplot(2,1,1)
plt.plot(a,decaying_sin,'-g')
plt.xticks(visible = False)
plt.title('Sin Graph Approaching Zero')
plt.legend()

plt.subplot(2,1,2)
plt.plot(a,sin,'o-')
plt.title('Sin Graph')

plt.show()
```

## 여러 그래프 그리기 2 : subplots

```python
fig , ax = plt.subplots(2,1) # figure와 ax 를 같이 설정해줘야 함
ax[0].plot(a,decaying_sin,'g-')# ax[행,열] 이지만 지금은 2개 그래프이기에 인덱스 하나로 표현 가능 
ax[1].plot(a,sin,'o-')# ax[행,열]

plt.show()
```

- *subplot*과 *subplots*는 *state-base* 와 *object-oriented* 방식의 차이

## 두 종류의 그래프 하나에 담기 : 바 그래프 & 라인 그래프

아래의 데이터는 임의로 만든 ABC 라는 가상 국가의 1951년 부터 1960년까지 집 거래량과 연 평균 집값
- 연 평균 집값
- 총 거래량 
- 기간 

```python
periods = np.arange(1951,1961)# 연도 표기
housing_prices = np.array([15,25,31,24,22,27,40,55,60,59])
num_trading = np.array([6,6.6,7,4.5,4,6,9.5,11,20,18])
```

```python

fig, ax1 = plt.subplots(figsize=(9,5))
# 집값은 라인으로 표기 
ax1.plot(periods,housing_prices,'o-',linewidth =1 , alpha = 0.8 , label = 'housing_price')
ax1.set_xlabel('Year',fontsize = 13)
ax1.set_ylabel('Price($000)',fontsize = 13)
ax1.set_ylim(0,70)

# 총 거래량은 box로 표기 
ax2 = ax1.twinx()# twinx는 x축은 공유하지만 y축은 공유하지 않게 함 
ax2.bar(periods,num_trading, color = 'coral',alpha = 0.3 , label ='#houses traded') # alpha = 투명도 
ax2.set_ylim(0,22)

ax1.legend(loc='upper left')
ax2.legend(loc = 'upper right')

plt.tight_layout()
plt.show()
```

```python
subplot과 subplots는 state-base 와 object-oriented 방식의 차이
```
