```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data = pd.read_excel('D:/KTdata/boston.xls')
housing_data = pd.DataFrame(housing_data)
housing_data.describe()
```

```python
# 파이차트 실습
bins = [0,20,40,60,80,100]
names = ['G1','G2','G3','G4','G5']
housing_data['AGE_G5'] = pd.cut(housing_data['AGE'],bins,labels = names)
housing_data.keys()
```

AGE 는 오래된 주택비율을 의미하므로 AGE G5는 오래된 주택비율 등급을 G1(0%)~ G5(100%) 추가

```python
#파이차트 생성을 위해 빈도 table 생성
pie_df_G5 = pd.DataFrame(housing_data['AGE_G5'].value_counts())
# pie 차트 생성시 범주 별 크기에 따른 sorting 후 plotting
pie_df_G5
```

```python
#파이차트 생성 
pie_df_G5.plot.pie(y="AGE_G5", figsize = (10,10) , colors= sns.color_palette('hls', len(pie_df_G5)))
plt.show()
```
