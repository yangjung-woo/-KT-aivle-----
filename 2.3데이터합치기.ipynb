## 데이터프레임 Merge

```python
import pandas as pd
```

```python
data1 = {
    '학번': [1,2,3,4],
    '이름': ['양정우','김철수','둘리','신창섭'],
    '학과': ['컴공','철학','물리','신학']
}

data2 ={
    '학번':[1,2,4,5],
    '학년':[2,4,3,1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}
df1 =pd.DataFrame(data1)
df2 =pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# inner Join 사용 
inner_df = pd.merge(df1,df2,on="학번",how="inner") # 병합 기준 on = "학년"
display(inner_df)
```

```python
# Outer Join 사용 
outer_df = pd.merge(df1,df2,on="학번",how="outer") # 병합 기준 on = "학년"
display(outer_df)
```

### Left and Right Join

```python
left_df = pd.merge(df1,df2,on="학번",how="left") # df1 기준으로 병합
display(left_df)
```

```python
right_df = pd.merge(df1,df2,on="학번",how="right")#  df2 기준으로 병합
display(right_df)
```

### (merge) 응용: 현실 데ㅣ터를 다룰때 생길만 한 경우 

 1 같은 의미의 다른 컬럼 명을 병합할 때 

```python
data1 = {
    '학번': [1,2,3,4],
    '이름': ['양정우','김철수','둘리','신창섭'],
    '학과': ['컴공','철학','물리','신학']
}

data2 ={
    '학생고유번호':[1,2,4,5],
    '학년':[2,4,3,1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}
# 학번 = 학생고유번호 인 경우 의미는 같지만 컬럼명이 다른 경우 
df1 =pd.DataFrame(data1)
df2 =pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
case1_merge_df = pd.merge(df1,df2,left_on="학번",right_on="학생고유번호",how="inner")#df1, df2 inner join하되 의미는 같지만 명이 다른 경우 
display(case1_merge_df)
```

```python
# 두 개의 공통 키 중 하나 삭제
case1_merge_df.drop('학생고유번호',axis=1,inplace=True) 
# axis=1 : 열 삭제 , Inplace: drop한 결과를 해당 변수에 반영
```

#### 2 공유하는 컬럼이 없는 경우
> index로 병합 가능 

```python
data1 = {
    '학번': [1,2,3,4],
    '이름': ['양정우','김철수','둘리','신창섭'],
    '학과': ['컴공','철학','물리','신학']
}

data2 ={
    '학년':[2,4,3,1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}
# 학번 = 학생고유번호 인 경우 의미는 같지만 컬럼명이 다른 경우 
df1 =pd.DataFrame(data1)
df2 =pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# 공통되는 값이 없는 경우 index를 기준으로 병합 할 순 있긴함 (단 순서를 잘 지켜야 함 )
case2_merge_df= pd.merge(df1,df2,left_index=True,right_index=True,how='inner')
display(case2_merge_df)
```

```python
# 개선판 df1은 학번을 기준으로 , df2는 index 를 기준으로 inner join 
case2_merge_df= pd.merge(df1,df2,left_on='학번',right_index=True,how='inner')
display(case2_merge_df)
```

```python
# pandas가 아닌 dataframe이 제공하는 join : 양쪽 index 를 기준으로 병합
df1.join(df2,how='inner')
```

### 데이터 연결 Pandas , concat

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.randint(0,9,(3,2)), index = ['a','b','c'], columns= ['one','two'])
df2 = pd.DataFrame(np.random.randint(0,9,(2,2)), index = ['a','b'], columns= ['three','four'])
display(df1)
display(df2)
```

```python
#df1과 df2 concat으로 연결 
result_df = pd.concat([df1,df2] ,axis=1)# axis=1 : 열 방향으로 연결 
display(result_df)
```

```python
#outerjoin으로도 concat과 가튼 결과 확인 가능 
merge_df = pd.merge(df1,df2,left_index=True, right_index=True,how='outer')
print(merge_df)
```

```python
result_df = pd.concat([df1,df2] ,axis=0)# axis=0 : 행 방향으로 연결 
display(result_df)
```
