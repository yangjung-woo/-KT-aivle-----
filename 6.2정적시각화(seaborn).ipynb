## 정적 시각화 실습 - seaborn 활용

```python
#필요 라이브러리 로딩
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
display(df.head())
```

## seaborn의 scatterplot , regplot , 그리고 Implot
산점도는 수치형 변수간의 관계를 유용하게 파악하기 위한 시각화 방법
seaborn의 scatterplot , regplot , implot 모두 산점도에 기반한다는 공통점이 있음 

### 산점도: scatterplot

```python
# 스카터 플롯은 총 2개으 수치형 변수 x와 y 가 필요함

sns.scatterplot(x='sepal_length',y ='petal_length',data=df)
plt.show()
```

scatterplot의 hue 옵션

```python
# hue 인수의 값을 설정해준다면, 종류를 색상으로 나타낼 수 있음 
# 붗꽃 종류별로 다른 생상 표현 : hue = 'sepcies'
sns.scatterplot(x='sepal_length',y ='petal_length',hue='species',data=df)
plt.show()
```

### 산점도 + 추세선 : regplot
regplot 도 산점도에 기반함
scatterplot과 다른 점이 있다면, regplot을 이용하면 추세선과 추세선의 신뢰구간을 넣을 수 있다

```python
# regplot 호출
sns.regplot(x='sepal_length',y='petal_length',data=df) # hue를 허용하지 않음
plt.show()
```

### 범주별 산점도 + 추세선: lmplot

```python
sns.lmplot(x='sepal_length',y='petal_length',hue='species',data=df)
plt.show()
```
