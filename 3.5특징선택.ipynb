```python
import numpy as np
import pandas as pd
```

```python
import sklearn.datasets as d
# breast_cancer 데이터 셋 로드
x = d.load_breast_cancer()
cancer = pd.DataFrame(data = x.data, columns = x.feature_names)
cancer['target'] = x.target


cancer.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train,Y_test = train_test_split(cancer.drop(['dignosis'] , axis=1 ), cancer[['diagnosis']], random_state = 1)

print(X_train.shape)
print(X_test.shape) # 훈련용 : 평가용 데이터 셋 나누기 : 3:1
```

```python
train_df = pd.merge(X_train , Y_train, left_index=True , right_index=True, how = 'inner')
train_df # 독릭 종속 변수 훈련데이터 통합
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig , axes = plt.subplots(10,3,figsize=((20,40)))
axes = axes.flatten()
for i in range(30):
    sns.histplot(data=train_df, x=train_df.columns[i], hue='diagnosis', multiple='layer',ax=axes[i])
plt.show()
```
