
### groupby 

```python
import pandas as pd
import numpy as np
```

```python
df =pd.DataFrame ({
    '학번': [1,2,3,4],
    '이름': ['양정우','김철수','둘리','신창섭'],
    '학과': ['컴공','철학','물리','신학'],
    '학점': [1.5, 2.0, 2.5, 3.5]

})
display(df)
```

```python
df_dept = df.groupby('학과')# 학과를 기준으로 그룹화  
print(df_dept) # 
```

### 1단계 그룹핑

```python
#series에 대한 1단계 그룹팡
dept = df['학점'].groupby(df['학과']) # df['컬럼명']의 경우 series 로 반환
print(dept)
```

```python
# 학과 평균 성적 출력 
dept.mean()
```

```python
# 학과 표준편차 출력 
dept.std()
```

```python
#group별 크기 출력
group_size = dept.size()
print(group_size)
```

```python
#컴공 인원의 데이터만 따로 출력
computer = dept.get_group('컴공')
print(computer)
```

이처럼 *GroupBy* 객체의 그룹연산 메서드는 다양합니다. 자주 사용되는 메서드
- *mean* , median , min , max , std: 그룹 데이터의 평균, 중앙값, 최소값, 최대값, 표준편차
- size , count : 그룹 데이터의 갯수 
- describe: 위 집계 연산과 1사분위, 3사분위값을 포함하여 데이터프레임으로 나타냄
- sum , quantile: 그룹 데이터의 합계, 사분위수 
- agg, aggregate , apply : 원하는 그룹연산 함수를 만들어  agg, aggregate , apply 에 전달하여 적용할 수 있음

2단계 그룹핑

```python
#그룹별 기술통계량 도출
dept.describe()
```

```python
# 더 많은 데이터로 실습진행
df2 = pd.DataFrame({
    '학과':['화학','수학','화학','수학'],
    '이름':['앤디','수빈','무형','알리샤'],
    '학년':[1,2,3,2],
    '학점':[3,4.2,3.1,4.5]
})

#df1과 df2 행 방향 연결 
concat_df = pd.concat([df,df2], axis = 0)
display(concat_df)
```

```python
concat_df.reset_index(inplace=True, drop=True)# index 정렬
display(concat_df)
```

```python
#하과와 학년별로 그룹핑 
dept_and_year = concat_df.groupby([concat_df['학과'] , concat_df['학년']])
print(dept_and_year)
```

```python
dept_and_year.describe()
```

```python
#2단계 그룹화 인덱스 확인
dept_and_year.mean().index
```

apply함수
- 사용자 정의 함수를 적용해야 할때 사용

```python
# 학과별 우수 성적자 3명 출력 함수
def top3_dept_scorer(df):
    return df.sort_values(by='학점',ascending=False)[:3]# 학점 상위 3명만 반환   
```

```python
# 학과 그룹바이 객체 생성 
dept = concat_df.groupby('학과')
```

```python
# 사용자 정의 함수 적용 
dept.apply(top3_dept_scorer)
```

```python
# 학과별 우수 성적자 2명 출력 함수
def top2_scorer(series):
    return series.sort_values(ascending=False)[:2]# 학점 상위 2명만 반환   
```

```python
dept['학점'].apply(top2_scorer)
```

```python
# 멀티인덱스의 경우 데이터프레임으로 보면 정렬된 뷰를 볼수 있다 
pd.DataFrame(dept['학점'].apply(top2_scorer))
```

lambda식 활용

```python
# 새로운 데이터프레임 생성 
df = pd.DataFrame({
    'key': ['A','B','B','C','A','C','A','B','C','A'],
    'data': [1,2,2,1,3,8,2,5,3,6]
})
```

```python
#key별로 그룹화를 진행후 data 값의 크기별로 오름차순 정렬 
df.groupby('key').apply(lambda x:x.sort_values(by='data'))
```

```python
# 위에서 apply 함수를 이용하여 진행한 학과별 성적 탑 3명 학생 출력을 lambda로 진행
dept = concat_df.groupby('학과')
```

```python
dept.apply(lambda x: x.sort_values(by='학점',ascending=True)[:3])
```

groupby와 재귀함수 

```python
# 1차 그룹핑 재귀
for dept, group in concat_df.groupby(concat_df['학과']):
    print('학과 : {}'.format(dept))
    display(group)
```

```python
#2차 그룹핑 재귀 
for(dept,year), group in concat_df.groupby([concat_df['학과'], concat_df['학년']]):# 학과와 학년으로 그룹바이 진행 
    print('-'*25)
    print('학과: ',dept)
    print('학년: ',year)
    display(group)
```
